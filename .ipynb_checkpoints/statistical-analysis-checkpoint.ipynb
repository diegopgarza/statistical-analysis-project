```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('train.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
# Limpiemos el df. La columna ID es irrelevante.

df.drop(columns='Id', inplace=True)
```

```python
df.head()
```

```python
# Para un analisis mas exacto, necesitamos normalizar nuestra variable SalePrice:

df['SalePrice_Norm'] = np.log(df['SalePrice'])
df[['SalePrice', 'SalePrice_Norm']]
```

```python
sns.histplot(df['SalePrice_Norm']);
```

```python
null_percentage = df.isnull().sum() * 100 / df.shape[0]
missing_values = pd.DataFrame({'column_name': df.columns, 'null_percentage': null_percentage})
missing_values = missing_values[missing_values.null_percentage > 80].round(2)
col_to_drop = list(missing_values.column_name)
df.drop(columns = col_to_drop, inplace = True)
df.head(3)
```

```python
# Separamos variables categoricas y numericas

df_cat = df.select_dtypes(include=['object','category'])
df_num = df.select_dtypes(include='number')
```

```python
df_cat.columns
```

```python
anova_categoricas = ols('SalePrice_Norm ~ MSZoning + Street +LotShape + LandContour + Utilities + LotConfig + LandSlope + Neighborhood+Condition1+Condition2+BldgType+HouseStyle+RoofStyle+RoofMatl+Exterior1st+Exterior2nd+MasVnrType+ExterQual+ExterCond+Foundation+BsmtQual+BsmtCond+BsmtExposure+BsmtFinType1+BsmtFinType2+Heating+HeatingQC+CentralAir+Electrical+KitchenQual+Functional+FireplaceQu+GarageType+GarageFinish+GarageQual+GarageCond+PavedDrive+SaleType+SaleCondition', data = df).fit()
sm.stats.anova_lm(anova_categoricas).head()
```

```python
# Para ir disminuyendo nuestras columnas, nos quedaremos solo con las que tienen una confianza del 100%.

confianza_categoricas = round((1 - sm.stats.anova_lm(anova_categoricas)[['PR(>F)']].sort_values(by = 'PR(>F)')).rename(columns = {'PR(>F)':'Confianza'})*100, 2)
confianza_categoricas[confianza_categoricas['Confianza'] == 100]
```

```python
# Eliminare MSZoning, ya que Neighborhood incluye la Zona.

columnas_relevantes_categoricas = ['Neighborhood','LotShape','BldgType','Exterior1st','ExterQual','RoofStyle','LandContour','MasVnrType','MasVnrType','BsmtExposure','RoofMatl',
                                   'BsmtQual','HouseStyle','KitchenQual','SaleCondition','Condition2','LotConfig','BsmtFinType1','Condition1','Exterior2nd']
columnas_relevantes_categoricas
```

```python
for cat in columnas_relevantes_categoricas:
    print("\n",cat)
    print(df[cat].value_counts())
    print((df[cat].value_counts()/len(df[cat])))
```

```python
categoricas_dummies = pd.get_dummies(df_cat, columns = ['Neighborhood'])
categoricas_dummies
```

```python
# Yo pensaria que la variable mas importante deberia ser 'Neighborhood'

X = sm.add_constant(pd.get_dummies(df['Neighborhood']))
y = df['SalePrice_Norm']
model = sm.OLS(y, X).fit()
print_modelo1 = model.summary()
print(print_modelo1)
```

```python
# Para verificar lo anterior, nuevamente usaremos el sentido comÃºn, y veremos tambiÃ©n 
# por intuiciÃ³n si existe alguna de esas variables que tenga mÃ¡s peso en el valor del precio:

X = sm.add_constant(pd.get_dummies(df['SaleCondition']))
y = df['SalePrice_Norm']
model = sm.OLS(y, X).fit()
print_modelo1 = model.summary()
print(print_modelo1)
```

```python
X = sm.add_constant(pd.get_dummies(df['ExterQual']))
y = df['SalePrice_Norm']
model = sm.OLS(y, X).fit()
print_modelo1 = model.summary()
print(print_modelo1)
```

```python
# Utilizaremos aquellas variables que tengan mayor correlacion con 'SalePrice'. 
# Definiremos un 0.5 como numero de corte:

correlation = df_num.corr()
df_num_corr = correlation[(correlation['SalePrice_Norm']>=0.5)|(correlation['SalePrice_Norm']<=-0.5)].sort_values(by='SalePrice_Norm', ascending=False)
col_num = set(df_num_corr.index)
col_totales = set(correlation.columns)
col_to_del = list(col_totales - col_num)
df_num_corr.drop(columns = col_to_del, inplace=True)
df_num_corr.SalePrice_Norm
```
